**For greater precision when filtering the data, we create a new column with the unified date and drop the innecesary column**

```python
# Crear una columna 'fecha' con el año, mes y día
pollution_2020['fecha'] = pd.to_datetime(pollution_2020[['ano', 'mes', 'dia']].astype(str).agg('-'.join, axis=1), errors='coerce', format='%Y-%m-%d')
pollution_2020.drop(["ano", "mes", "dia"], axis=1, inplace=True)

weather_2020['fecha'] = pd.to_datetime(weather_2020[['ano', 'mes', 'dia']].astype(str).agg('-'.join, axis=1), errors='coerce', format='%Y-%m-%d')
weather_2020.drop(["ano", "mes", "dia"], axis=1, inplace=True)
```

```python
pollution_2021['fecha'] = pd.to_datetime(pollution_2021[['ano', 'mes', 'dia']].astype(str).agg('-'.join, axis=1), errors='coerce', format='%Y-%m-%d')
pollution_2021.drop(["ano", "mes", "dia"], axis=1, inplace=True)

weather_2021['fecha'] = pd.to_datetime(weather_2021[['ano', 'mes', 'dia']].astype(str).agg('-'.join, axis=1), errors='coerce', format='%Y-%m-%d')
weather_2021.drop(["ano", "mes", "dia"], axis=1, inplace=True)
```

```python
pollution_2022['fecha'] = pd.to_datetime(pollution_2022[['ano', 'mes', 'dia']].astype(str).agg('-'.join, axis=1), errors='coerce', format='%Y-%m-%d')
pollution_2022.drop(["ano", "mes", "dia"], axis=1, inplace=True)

weather_2022['fecha'] = pd.to_datetime(weather_2022[['ano', 'mes', 'dia']].astype(str).agg('-'.join, axis=1), errors='coerce', format='%Y-%m-%d')
weather_2022.drop(["ano", "mes", "dia"], axis=1, inplace=True)
```

```python
pollution_2023['fecha'] = pd.to_datetime(pollution_2023[['ano', 'mes', 'dia']].astype(str).agg('-'.join, axis=1), errors='coerce', format='%Y-%m-%d')
pollution_2023.drop(["ano", "mes", "dia"], axis=1, inplace=True)

weather_2023['fecha'] = pd.to_datetime(weather_2023[['ano', 'mes', 'dia']].astype(str).agg('-'.join, axis=1), errors='coerce', format='%Y-%m-%d')
weather_2023.drop(["ano", "mes", "dia"], axis=1, inplace=True)
```

```python
#Eliminate columns that starts with 'v' and innecesary columns
pollution_2020.drop(columns=['punto_muestreo'] + [col for col in pollution_2020.columns if 'v' in col], inplace=True)
weather_2020.drop(columns= ['punto_muestreo'] + [col for col in weather_2020.columns if 'v' in col], inplace= True)
```

```python
pollution_2021.drop(columns=['punto_muestreo'] + [col for col in pollution_2021.columns if 'v' in col], inplace=True)
weather_2021.drop(columns= ['punto_muestreo'] + [col for col in weather_2021.columns if 'v' in col], inplace= True)
```

```python
pollution_2022.drop(columns=['punto_muestreo'] + [col for col in pollution_2022.columns if 'v' in col], inplace=True)
weather_2022.drop(columns= ['punto_muestreo'] + [col for col in weather_2022.columns if 'v' in col], inplace= True)
```

```python
pollution_2023.drop(columns=['punto_muestreo'] + [col for col in pollution_2023.columns if 'v' in col], inplace=True)
weather_2023.drop(columns= ['punto_muestreo'] + [col for col in weather_2023.columns if 'v' in col], inplace= True)
```
