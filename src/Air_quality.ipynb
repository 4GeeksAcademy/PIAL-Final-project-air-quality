```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pollution_2020 = pd.read_csv('../data/raw/pollution data/2020.csv', sep = ';')
pollution_2020.head()
```

```python
pollution_2020.drop(["provincia", "estacion", "punto_muestreo"], axis=1, inplace=True)
pollution_2020.head()
```

```python
# Iterar sobre todas las columnas "object" y convertir a tipo float con reemplazo de comas por puntos
for columna in pollution_2020:
    if pollution_2020[columna].dtype == 'object':
        try:
            pollution_2020[columna] = pollution_2020[columna].str.replace(',' , '.').astype(float)
        except ValueError:
            pass
```

```python
pollution_2020.head()
```

```python
# Crear una columna 'fecha' con el aÃ±o, mes y dÃ­a
pollution_2020['fecha'] = pd.to_datetime(pollution_2020[['ano', 'mes', 'dia']].astype(str).agg('-'.join, axis=1) + ' 00:00', errors='coerce', format='%Y-%m-%d %H:%M')
pollution_2020.drop(["ano", "mes", "dia"], axis=1, inplace=True)
pollution_2020.head()
```

```python
# Convertir las columnas de hora en formato de tiempo
horas = pollution_2020.filter(like='h').columns
```
